```python
import pandas as pd
import sqlite3 as lite
```

```python
# Jupyer notebook을 server IP 접속
# 이 노트북이 열려있는 위치를 기준이로 접속하고자 하는 데이터베이스 파일이 어디에있는지 파악
db = lite.connect('./products.db')
```

```python
#시행하고자 하는 query
query = "SELECT * FROM Customers;"

# make cursor
cur = db.cursor() #cursor 생성
# send query to cursor
cur.execute(query)
# fetch results
cur.fetchall()[0]
```

### 하지만 이렇게 cursor만을 사용하면 새로운 query를 fetch할때 마다 execute, fetchall해야함으로 very inefficient

```python
# using Pandas. Arguments are 1. the query and 2. the target database
df = pd.read_sql(query, db)
# fetches the first 5
df.head()
```

```python
query = """
SELECT * FROM sqlite_master;
"""

pd.options.display.max_colwidth = 300
pd.read_sql(query, db)
# .tables를 했을때와 같은 결과
```

```python
# Query table names and each table's schema
query = """
SELECT tbl_name, sql FROM sqlite_master;
"""
pd.read_sql(query, db)
```

```python
query = """
SELECT * FROM Customers;
"""
df = pd.read_sql(query, db)
fran = df['Country'] == "France"
germ = df['Country'] == "Germany"
df[fran|germ]
```

```python
query = """
SELECT * FROM Customers
WHERE Country in ('Germany', 'France');
"""
pd.read_sql(query, db)
```

## Filter by Query

```python
query = """
SELECT * FROM Customers
WHERE Country='Germany' OR City='Paris';
"""

pd.read_sql(query, db)
```

## Filter by Pandas

```python
query = """
SELECT * FROM Customers;
"""
df = pd.read_sql(query, db)
df[(df["Country"]=="Germany") | (df["City"]=="France")]
```
