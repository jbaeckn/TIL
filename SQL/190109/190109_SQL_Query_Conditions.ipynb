## Homework Review
OrderDetails에서 Quantity가 40개 이상이며, Customers의 CustomerName이 Ernst나 Stop을 포함하는 전체 데이터를 선택하세요.

```python
pd.options.display.max_colwidth = 200
```

```python
query = '''
SELECT *
FROM Orders
WHERE
    OrderID IN(
        SELECT OrderID
        FROM OrderDetails
        WHERE
            Quantity >= 40
    )
    AND
    CustomerID IN(
        SELECT CustomerID
        FROM
            Customers
        WHERE
            CustomerName LIKE '%Ernst%'
            OR
            CustomerName LIKE '%Stop%'
    )
;
'''

pd.read_sql(query,db)
```

```python
query = '''
SELECT C.CustomerID, C.CustomerName, O.CustomerID, O.OrderID
FROM Customers C
    JOIN
        Orders O
    ON
        C.CustomerID = O.customerID
;
'''
#JOIN은 기본이 INNER JOIN이다.
pd.read_sql(query, db)
```

## Group By Query

```python
query = '''
SELECT *
FROM Orders;
'''
orders_df = pd.read_sql(query, db)
orders_df.head()
```

```python
#Using Pandas
date_group = orders_df.groupby('OrderDate')
date_group.get_group('1996-07-04')
```

```python
orders_df['OrderDate'].unique()
```

```python
daily_order_count = pd.DataFrame([
    {
        "OrderDate" : OrderDate,
        "Count" : len(date_group.get_group(OrderDate))
    } for OrderDate in orders_df["OrderDate"].unique()
    
])
daily_order_count
```

```python
query = '''
SELECT OrderDate, count(*)
FROM
    Orders
    GROUP BY
        OrderDate;
'''
pd.read_sql(query, db)
```
