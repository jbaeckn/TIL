```python
import pandas as pd
import sqlite3 as lite
```

```python
db = lite.connect('./hr.db')
```

## Question 1A
Select the names and job start dates of all employees who work for the department number 5

```python
query1a = '''
SELECT E.f_name, E.l_name, J.start_date, E.dep_id
FROM employees as E
INNER JOIN job_history as J
ON E.emp_id=J.empl_id
WHERE E.dep_id=5;
'''
pd.read_sql(query1a, db)
```

## Question 1B
Select the names, job start dates, and job titles of all employees who work for the department number 5

```python
query1b = '''
SELECT E.f_name, E.l_name, E.dep_id, JH.start_date, J.job_title
FROM employees as E
INNER JOIN job_history as JH
ON E.emp_id=JH.empl_id
INNER JOIN jobs as J
ON JH.jobs_id=J.job_ident
WHERE E.dep_id=5;
'''
pd.read_sql(query1b, db)
```

## Question 2A
Perform a Left Outer Join on the EMPLOYEES and DEPARTMENT tables and select employee id, last name, department id and department name for all employees

```python
query2a = '''
SELECT E.emp_id, E.l_name, E.dep_id, D.dep_name
FROM employees as E
LEFT OUTER JOIN departments as D
ON E.manager_id=D.manager_id;
'''
pd.read_sql(query2a, db)
```

## Question 2B
Re-write the query for 2A to limit the result set to include only the rows for employees born before 1980

```python
query2b = '''
SELECT E.emp_id, E.l_name, E.b_date, E.dep_id, D.dep_name
FROM employees as E
LEFT OUTER JOIN departments as D
ON E.manager_id=D.manager_id
WHERE SUBSTR(E.b_date, 7, 3) < '198';
'''
pd.read_sql(query2b, db)
```

## Question 2C
Re-write the query for 2A to have the result set include all the employees but department names for only the employees who were born before 1980

```python
query2c = '''
SELECT E.emp_id, E.l_name, E.dep_id, D.dep_name
FROM employees as E
LEFT OUTER JOIN departments as D
ON E.manager_id=D.manager_id
AND SUBSTR(E.b_date, 7, 3) < '198';
'''
pd.read_sql(query2c, db)

#Lesson Learned
#When you join two tables, you can set more than one conditions, for example, join 'on' one condition 'and' on another
```

## Question 3A
Perform a Full Join on the EMPLOYEES and DEPARTMENT tables and select the First name, Last name and Department name of all employees

```python
query3a = '''
SELECT E.emp_id, E.f_name, E.l_name, D.dep_name
FROM employees as E
LEFT OUTER JOIN departments as D
ON E.manager_id=D.manager_id

UNION

SELECT E.emp_id, E.f_name, E.l_name, D.dep_name
FROM departments as D
LEFT OUTER JOIN employees as E
ON E.manager_id=D.manager_id

'''
pd.read_sql(query3a, db)
```

## Question 3B
Re-write Query 3A to have the result set include all employee names but department id and department names only for male employees

```python
query3b = '''
SELECT E.emp_id, E.f_name, E.l_name, D.dep_name
FROM employees as E
LEFT OUTER JOIN departments as D
ON E.manager_id=D.manager_id AND E.sex='M'

UNION

SELECT E.emp_id, E.f_name, E.l_name, D.dep_name
FROM departments as D
LEFT OUTER JOIN employees as E
ON E.manager_id=D.manager_id AND E.sex='M'

'''
pd.read_sql(query3b, db)
```
